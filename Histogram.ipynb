```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# number of data points
n = 1000
# generate log normal distribution
data = np.exp(np.random.randn(n)/2)
plt.plot(data, 'ko')
```

```python
plt.hist(data)
```

```python
# define the number of bins to see the shape of the distibution
k=40
plt.hist(data,bins=k)
```

```python
y,x = np.histogram(data,bins=k) #x is equal to the number of boundaries for the bins

# we need to check the bins' centers
xx = (x[1:]+x[:-1])/2

# line histogram
plt.plot(xx,y)
plt.title('Histogram with ' +str(k)+ ' bins')
```
